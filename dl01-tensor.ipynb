```python
x = np.array([[1.0, 0.0]]) # 입력 값, 샘플 1개, 특성 2개

dense = keras.layers.Dense(
    units=1, # 해당 층의 뉴런 갯수(출력 노드) 1개
    activation="sigmoid",
    input_shape = (2,) # 입력되는 특성 수가 2개
) # 층 1개 생성
```

# Keras : 고수준 API
- Sequential : 층을 순서대로 쌓겠다.
- Dense : 퍼셉트론 여러개를 하나의 층으로 묶은 연결층

1. 모델을 정의한다
2. 레이어 블록을 쌓는다.
3. 컴파일한다
4. 학습한다.
5. 평가한다.

```python

output = dense(x) 

# 이제 가중치가 정상적으로 생성되었으므로 에러 없이 잘 뽑힙니다.
w, b = dense.get_weights()  

print("w shape:", w.shape)
print("b shape:", b.shape)
```

```python
# 행렬 곱셈을 하는 층을 만든 것
# 행렬 곱의 shape (1, 2) * (2, 1) = (1, 1)
# (2,1) * (1, 2) = (2, 2)
w @ x
```
