```python
import tensorflow as tf
import keras
from keras import layers, models
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. CIFAR-10 데이터 세트 로드 및 전처리
(train_input, train_target), (test_input, test_target) = keras.datasets.cifar10.load_data()

train_scaled = train_input / 255.0
test_scaled = test_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42
)

# 2. 업그레이드된 CNN 모델 설계
model = models.Sequential()
model.add(layers.Input(shape=(32, 32, 3)))

# [블록 1] 필터 32개 + 배치 정규화
model.add(layers.Conv2D(32, kernel_size=3, activation='relu', padding='same'))
model.add(layers.BatchNormalization()) # 과대적합 방지 및 학습 가속화!
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Dropout(0.25))        # 합성곱 단계에서의 규제 추가

# [블록 2] 필터 64개 + 배치 정규화
model.add(layers.Conv2D(64, kernel_size=3, activation='relu', padding='same'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Dropout(0.25))

# [블록 3] 필터 128개 + 배치 정규화
model.add(layers.Conv2D(128, kernel_size=3, activation='relu', padding='same'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Dropout(0.4))         # 깊은 층일수록 드롭아웃 강하게

# 분류층 데이터 직렬화
model.add(layers.Flatten()) 

# 은닉층 (Dense)
model.add(layers.Dense(256, activation='relu'))
model.add(layers.BatchNormalization()) # 밀집층 뒤에도 정규화 적용
model.add(layers.Dropout(0.5))         # 최종 과대적합 제어를 위해 0.5로 상향

# 출력층
model.add(layers.Dense(10, activation='softmax'))

model.summary()

# 3. 모델 컴파일 (Adam의 학습률을 0.001에서 0.0005로 반으로 줄여 더 꼼꼼하게 정밀 학습)
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 4. 콜백 설정 (더 정밀해졌으므로 무작위 진동을 견디도록 patience를 5로 확대)
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-cifar10-final.keras', save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)

# 5. 모델 학습 진행
history = model.fit(
    train_scaled, train_target, 
    epochs=40,            # 더 꼼꼼히 배우므로 에포크를 40으로 여유 있게 설정
    batch_size=64, 
    validation_data=(val_scaled, val_target),
    callbacks=[checkpoint_cb, early_stopping_cb]
)

# 6. 학습 결과 시각화
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.title('CIFAR-10 Upgraded CNN Training Loss')
plt.show()

# 7. 최종 테스트 데이터 평가
print("\n=== 최종 최적화 모델 테스트 데이터 평가 ===")
test_loss, test_acc = model.evaluate(test_scaled, test_target, verbose=2)
print(f"테스트 정확도(Test Accuracy): {test_acc:.4f}")
```
